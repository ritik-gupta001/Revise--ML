```python
# extracting episode and saving it into new column
def extract_episode(txt):
    check=False
    data=""
    for i in txt:
        if i ==')':
            break
        if check==True:
            data+=i
        if i =='(':
            check=True
    return data 
        
```

```python
df["Episodes"]=df['Title'].apply(extract_episode)
```

```python
df["Episodes"]=df["Episodes"].str.replace(" eps","")
```

```python
df["Episodes"]=df["Episodes"].astype(int)
```

```python
# extracting time stamp and saving it into new column
```

```python
def time_extract(txt):
    check=False
    data=""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range (i+1,i+20):
                data+=txt[j]
            return data
                
    
```

```python
df["Date"]=df["Title"].apply(time_extract)
```
