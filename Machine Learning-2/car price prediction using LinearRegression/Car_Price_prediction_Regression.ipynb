```python
 X=df.drop(columns=['price'],axis=1)
y=df['price']
```

```python
X.head()
```

```python
y.head()
```

```python
X_one_encoded=pd.get_dummies(X,columns=['model','transmission','fuelType'],drop_first=True)
X_one_encoded.head()
```

```python
X_one_encoded=X_one_encoded.astype(int)
X_one_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()
X_one_encoded[numerical_cols] = scaler.fit_transform(X_one_encoded[numerical_cols])
```

```python
X_one_encoded.head()
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

Xlabel = X.copy()  # make a safe copy
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    Xlabel[col] = le.fit_transform(Xlabel[col].astype(str))  # Convert to string in case of nulls
    label_encoders[col] = le
```

```python
Xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(Xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])

## labelencoding
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_one_encoded,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
print("Adjusted R Score:", adjusted_r2)
```
